```python
import csv 
import requests 
from bs4 import BeautifulSoup

#url = 'https://www.udemy.com/'
url = 'https://www.w3resource.com/'

res = requests.get(url=url)
# res

soup = BeautifulSoup(res.text, 'html.parser')
# soup

title_header = soup.find('title')

if title_header is None:
    print("No Title present for :", url)
else:
    print("Title is present for :", url)
    print(' ')
    print(title_header)
    print(title_header.text)
```
